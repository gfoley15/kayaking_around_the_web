```python
# Import dependencies
import requests
from pymongo import MongoClient
import time
import json
import pandas as pd
from pprint import pprint
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)

# assign the met database to a variable name
db = mongo['kayak']

# assign the collection to a variable
flight_search = db['flight_search']
```

```python
flight_search = list(flight_search.find({}))
```

```python
# Create a data frame for flight search df
flight_search_df = pd.DataFrame(flight_search)
```

```python
# Split the outbound_route to get destination IATA code 
flight_search_df[['outbound_airport', 'IATA']] = flight_search_df['outbound_route'].str.split('-', expand=True)
```

```python
# Review flight search df types
flight_search_df.dtypes
```

```python
# Replace '+1' in all time columns
flight_search_df['outbound_departure_time'] = flight_search_df['outbound_departure_time'].str.replace('+1','')
flight_search_df['outbound_arrival_time'] = flight_search_df['outbound_arrival_time'].str.replace('+1','')
flight_search_df['return_departure_time'] = flight_search_df['return_departure_time'].str.replace('+1','')
flight_search_df['return_arrival_time'] = flight_search_df['return_arrival_time'].str.replace('+1','')
```

```python
# Show the DataFrame
flight_search_df
```

```python
# Export for collaboration purposes
flight_search_df.to_csv('../Output/flight_search.csv',index=False) 
```
